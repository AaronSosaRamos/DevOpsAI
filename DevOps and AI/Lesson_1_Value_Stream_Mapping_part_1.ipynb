#How to develop a Value Stream Mapping diagram

image.png

```python
import matplotlib.pyplot as plt

# Define the processes and steps
processes = [
    'Supplier', 'Transport', 'Receiving', 'Inventory', 'Production',
    'Quality Control', 'Packaging', 'Distribution', 'Customer'
]

# Define the data related to each process (e.g., cycle time, lead time)
cycle_times = [1, 2, 3, 2, 5, 4, 2, 3, 1]  # Example cycle times in days
lead_times = [5, 2, 1, 1, 7, 3, 2, 3, 1]     # Example lead times in days

# Plotting the Value Stream Mapping diagram
plt.figure(figsize=(10, 6))
plt.barh(processes, cycle_times, color='skyblue', label='Cycle Time')
plt.barh(processes, lead_times, left=cycle_times, color='orange', label='Lead Time')

# Adding labels and titles
plt.xlabel('Days')
plt.title('Value Stream Mapping Diagram')
plt.legend()

# Show plot
plt.show()
```

#Time Metrics

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

# Define descriptive process step names and simulated data for various metrics
process_steps = [
    'Raw Material Acquisition',
    'Component Manufacturing',
    'Assembly',
    'Quality Control',
    'Packaging',
    'Warehouse Storage',
    'Distribution'
]

num_steps = len(process_steps)

# Simulated data for cycle times (in minutes)
cycle_times = np.random.randint(15, 45, size=num_steps)

# Simulated data for wait times (in minutes)
wait_times = np.random.randint(5, 15, size=num_steps)

# Simulated data for setup times (in minutes)
setup_times = np.random.randint(10, 20, size=num_steps)

# Simulated data for processing times (in minutes)
processing_times = np.random.randint(30, 60, size=num_steps)

# Simulated data for lead times (sum of cycle times and wait times)
lead_times = cycle_times + wait_times

# Plotting the metrics for each process step
plt.figure(figsize=(14, 10))

# Cycle Time vs. Lead Time Plot
plt.subplot(2, 2, 1)
plt.plot(process_steps, cycle_times, marker='o', label='Cycle Time', color='skyblue')
plt.plot(process_steps, lead_times, marker='s', label='Lead Time', color='orange')
plt.title('Cycle Time vs. Lead Time')
plt.xlabel('Process Steps')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)
plt.legend()

# Wait Time Plot
plt.subplot(2, 2, 2)
plt.bar(process_steps, wait_times, color='green')
plt.title('Wait Times')
plt.xlabel('Process Steps')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)

# Setup Time Plot
plt.subplot(2, 2, 3)
plt.bar(process_steps, setup_times, color='purple')
plt.title('Setup Times')
plt.xlabel('Process Steps')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)

# Processing Time Plot
plt.subplot(2, 2, 4)
plt.bar(process_steps, processing_times, color='red')
plt.title('Processing Times')
plt.xlabel('Process Steps')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```

#Flows

```python
import matplotlib.pyplot as plt

# Define process steps and data for material flow, information flow, and lead time ladder
process_steps = [
    'Raw Material Acquisition',
    'Component Manufacturing',
    'Assembly',
    'Quality Control',
    'Packaging',
    'Warehouse Storage',
    'Distribution'
]

material_flow = [100, 80, 70, 60, 50, 40, 30]  # Material quantity (e.g., units) at each step
information_flow = [50, 45, 40, 35, 30, 25, 20]  # Information flow (e.g., documents) at each step
lead_time = [5, 10, 8, 7, 6, 4, 3]  # Lead time (in days) at each step

# Plotting Material Flow
plt.figure(figsize=(10, 6))
plt.plot(process_steps, material_flow, marker='o', linestyle='-', color='blue')
plt.title('Material Flow in the Value Stream')
plt.xlabel('Process Steps')
plt.ylabel('Material Quantity')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Plotting Information Flow
plt.figure(figsize=(10, 6))
plt.plot(process_steps, information_flow, marker='s', linestyle='--', color='green')
plt.title('Information Flow in the Value Stream')
plt.xlabel('Process Steps')
plt.ylabel('Information Volume')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Plotting Lead Time Ladder
plt.figure(figsize=(10, 6))
plt.bar(process_steps, lead_time, color='orange')
plt.title('Lead Time Breakdown in the Value Stream')
plt.xlabel('Process Steps')
plt.ylabel('Lead Time (Days)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define process steps and simulated data for VSM metrics
process_steps = [
    'Step 1', 'Step 2', 'Step 3', 'Step 4', 'Step 5', 'Step 6', 'Step 7'
]
num_steps = len(process_steps)

# Simulated data for different VSM metrics (in minutes, units, etc.)
cycle_times = np.random.randint(20, 40, size=num_steps)
process_times = np.random.randint(15, 30, size=num_steps)
changeover_times = np.random.randint(5, 15, size=num_steps)
inventory_levels = np.random.randint(50, 200, size=num_steps)
value_added_times = np.random.randint(10, 20, size=num_steps)

# Plotting VSM metrics
plt.figure(figsize=(14, 12))

# Cycle Time Plot
plt.subplot(3, 2, 1)
plt.plot(process_steps, cycle_times, marker='o', linestyle='-', color='blue')
plt.title('Cycle Time across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)
plt.grid(True)

# Process Time Plot
plt.subplot(3, 2, 2)
plt.plot(process_steps, process_times, marker='s', linestyle='--', color='green')
plt.title('Process Time across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)
plt.grid(True)

# Changeover Time Plot
plt.subplot(3, 2, 3)
plt.bar(process_steps, changeover_times, color='orange')
plt.title('Changeover Time across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)
plt.grid(True)

# Inventory Levels Plot
plt.subplot(3, 2, 4)
plt.bar(process_steps, inventory_levels, color='purple')
plt.title('Inventory Levels across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Inventory (units)')
plt.xticks(rotation=45)
plt.grid(True)

# Value-Added Time Plot
plt.subplot(3, 2, 5)
plt.plot(process_steps, value_added_times, marker='*', linestyle='-', color='red')
plt.title('Value-Added Time across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)
plt.grid(True)

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```

#Quality Metrics (ISO 9000:2015 and ISO 9001:2015)

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

# Define descriptive process step names and simulated data for VSM metrics with quality metrics
process_steps = [
    'Raw Material Procurement',
    'Component Fabrication',
    'Assembly Line',
    'Quality Inspection',
    'Packaging',
    'Warehouse Management',
    'Product Distribution'
]
num_steps = len(process_steps)

# Simulated data for quality metrics
defect_rates = np.random.uniform(0.5, 5.0, size=num_steps)  # Defect rate per 100 units
customer_complaints = np.random.randint(0, 10, size=num_steps)  # Number of complaints
conformance_levels = np.random.uniform(70, 95, size=num_steps)  # Conformance percentage (%)
customer_satisfaction = np.random.uniform(70, 100, size=num_steps)  # Customer satisfaction (%)
process_capability = np.random.uniform(1.0, 1.5, size=num_steps)  # Process capability index (Cpk)

# Plotting Enhanced VSM with Quality Metrics
plt.figure(figsize=(16, 12))

# Defect Rate Plot
plt.subplot(3, 2, 1)
plt.plot(process_steps, defect_rates, marker='o', linestyle='-', color='blue')
plt.title('Defect Rate across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Defect Rate (per 100 units)')
plt.xticks(rotation=45)
plt.grid(True)

# Customer Complaints Plot
plt.subplot(3, 2, 2)
plt.bar(process_steps, customer_complaints, color='green')
plt.title('Customer Complaints across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45)
plt.grid(True)

# Conformance to Specifications Plot
plt.subplot(3, 2, 3)
plt.plot(process_steps, conformance_levels, marker='s', linestyle='--', color='orange')
plt.title('Conformance to Specifications across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Conformance (%)')
plt.xticks(rotation=45)
plt.grid(True)

# Customer Satisfaction Plot
plt.subplot(3, 2, 4)
plt.plot(process_steps, customer_satisfaction, marker='*', linestyle='-', color='purple')
plt.title('Customer Satisfaction across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Customer Satisfaction (%)')
plt.xticks(rotation=45)
plt.grid(True)

# Process Capability Plot
plt.subplot(3, 2, 5)
plt.bar(process_steps, process_capability, color='red')
plt.title('Process Capability across Process Steps')
plt.xlabel('Process Steps')
plt.ylabel('Process Capability (Cpk)')
plt.xticks(rotation=45)
plt.grid(True)

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```
