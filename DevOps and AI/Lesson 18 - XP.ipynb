#Simplicity
The principle of simplicity in Extreme Programming emphasizes writing the simplest code to achieve the desired functionality. This involves avoiding unnecessary complexity, keeping methods and classes concise, and focusing on clarity and readability.

```python
# Class representing a shopping cart
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item, quantity):
        self.items.append({'item': item, 'quantity': quantity})

    def calculate_total(self):
        total = 0
        for item in self.items:
            total += item['item'].price * item['quantity']
        return total

# Usage
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

item1 = Item('Laptop', 1000)
item2 = Item('Mouse', 20)

cart = ShoppingCart()
cart.add_item(item1, 1)
cart.add_item(item2, 2)

total_cost = cart.calculate_total()
print(f"Total cost of items in the cart: ${total_cost}")
```

#Feedback

XP promotes frequent feedback loops to ensure continuous improvement. This involves gathering feedback from customers, team members, and stakeholders early and often, which helps to adapt and refine the product.

```python
# Class representing a feedback system
class FeedbackSystem:
    def __init__(self):
        self.ratings = []

    def add_rating(self, rating):
        self.ratings.append(rating)

    def calculate_average_rating(self):
        if not self.ratings:
            return 0
        return sum(self.ratings) / len(self.ratings)

# Usage
feedback = FeedbackSystem()

# Simulate collecting feedback
feedback.add_rating(4)
feedback.add_rating(5)
feedback.add_rating(3)

average_rating = feedback.calculate_average_rating()
print(f"Average rating based on user feedback: {average_rating:.2f}")
```

#Communication


Effective communication among team members is crucial.

```python
# Class representing a team messaging system
class TeamMessagingSystem:
    def __init__(self):
        self.messages = []

    def send_message(self, sender, receiver, message):
        self.messages.append({'sender': sender, 'receiver': receiver, 'message': message})

    def display_messages(self):
        for message in self.messages:
            print(f"From {message['sender']} to {message['receiver']}: {message['message']}")

# Usage
messaging_system = TeamMessagingSystem()

# Simulate team communication
messaging_system.send_message('Alice', 'Bob', 'Lets discuss the project requirements.')
messaging_system.send_message('Bob', 'Alice', 'Sure, when are you available?')

messaging_system.display_messages()
```

#Courage

```python
# Original complex function
def complex_function(data):
    result = 0
    for item in data:
        if item > 0:
            result += item ** 2
        else:
            result -= item
    return result

# Refactored simpler function
def simplified_function(data):
    return sum(item ** 2 if item > 0 else -item for item in data)

# Usage
data = [1, -2, 3, -4, 5]
result_old = complex_function(data)
result_new = simplified_function(data)

print(f"Result (original): {result_old}")
print(f"Result (refactored): {result_new}")
```

#Respect

```python
# Class representing a team with members
class Team:
    def __init__(self, members):
        self.members = members

    def appreciate_member(self, member):
        print(f"Thank you, {member}, for your hard work and dedication!")

# Usage
team = Team(['Alice', 'Bob', 'Charlie'])

# Appreciate team members
for member in team.members:
    team.appreciate_member(member)
```
